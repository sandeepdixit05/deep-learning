# XOR using Multi-Layer Perceptron (MLP) model

We saw that perceptron network was not able to classify all the points for XOR data. 

Let us see how MLP architecture with single hidden layer performs on XOR data.

## Importing Packages and Modules

```python
import numpy as np
```

## Input & Output Data

```python
num_inputs = 4  # Number of data points
```

```python
input_array = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
])
print(input_array)
```

```python
output_array = np.array([0, 1, 1, 0])   # XOR output
output_array
```

## MLP Model


### Importing the Keras libraries and packages

```python
  from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

### Initialising the model

```python
classifier = Sequential()
```

### Adding Layers

```python
# Adding the input layer and the first hidden layer with 'relu' activation function.
classifier.add(Dense(input_shape = (2,), units=10, activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units=1, activation = 'sigmoid'))
```

### Model Summary

```python
classifier.summary()
```

### Model Compiling

```python
classifier.compile(optimizer = SGD(lr=0.9), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Fitting the model

```python
classifier.fit(input_array, output_array, epochs = 50)   # Loss reduces with epoch, but may take more epochs
```

## Model Prediction & Validation

```python
# The trained model is used to predict and evaluate the model, 
# with cut-off value of 0.5

y_pred = classifier.predict(input_array)
y_pred = (y_pred > 0.5)
```

```python
# Accuracy Score.
from sklearn.metrics import accuracy_score

print("Accuracy: {}".format(accuracy_score(output_array, y_pred)))                            
```

We observe that the model with single hidden layer is able to get an accuracy of 100% and able to classify all the points correctly.
